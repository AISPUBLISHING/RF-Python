```python
#Data cleaning
#missing value
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic =  pd.read_csv('titanic.csv')
```

```python
titanic.tail()
```

```python
titanic['Age'].isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace = True)
```

```python
titanic.tail()
```

```python
titanic['Age'].isnull().sum()
```

```python
titanic.to_csv('missing_values_replaced.csv', index = False)
```

```python
#conversion from catagorical to numeric values
titanic = pd.read_csv('after_removal_outliers.csv')
```

```python
titanic.head()
```

```python
pd.get_dummies(titanic['Sex'], prefix='Sex')
```

```python
sex = pd.get_dummies(titanic['Sex'], prefix = 'Sex')
embaraked = pd.get_dummies(titanic['Embarked'], prefix = 'Embarked')
Pclass = pd.get_dummies(titanic['Pclass'], prefix = 'Pclass')
```

```python
new_titanic = pd.concat([titanic, sex, embaraked, Pclass],axis=1)
```

```python
new_titanic.head()
```

```python
new_titanic.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Embarked', 'Ticket', 'Cabin'], axis=1, inplace = True)
```

```python
new_titanic.head()
```

```python
new_titanic.to_csv('final_dataset.csv', index=False)
```
