```python
#Matplotlib
import matplotlib.pyplot as plt
```

```python
y = titanic[titanic['Survived']==1].groupby('Sex')['Name'].count()
x = ['Females', 'Males']
plt.bar(x,y)
plt.title('Survival wrt Gender')
```

```python
y = titanic[titanic['Survived']==1].groupby('Pclass')['Name'].count()
x = ['Class 1', 'Class 2', 'Class 3']
plt.bar(x,y)
plt.title('Survival wrt Class')
```

```python
childs = titanic[(titanic['Survived']==1) & (titanic['Age']<=12)]['Name'].count()
adults = titanic[(titanic['Survived']==1) & (titanic['Age']>12)]['Name'].count()
```

```python
y = [childs, adults]
x = ['Childs', 'Adults']
plt.bar(x,y)
plt.title('Survival wrt age')
```

```python
single = titanic[(titanic['Survived']==0) & (titanic['SibSp']==0) & (titanic['Parch']==0)]['Name'].count()
fam = titanic['Name'].count() - single
plt.bar(['Singles', 'Families'],[single, fam])
plt.title('Died wrt Families')
```
