```python
def addition(num1, num2):
    return num1 + num2

def subtraction(num1, num2):
    return num1 - num2
```

```python
def take_input():
    num1 = int(input('Enter num 1 : '))
    num2 = int(input('Enter num 2 : '))
    return num1, num2
```

```python
def calculate(choice):
    if choice == 1:
        num1, num2 = take_input()
        addition(num1, num2)
    elif choice == 2:
        num1, num2 = take_input()
        subtraction(num1, num2)
    elif choice == 3:
        num1, num2 = take_input()
        multiplication(num1, num2)
    elif choice == 4:
        num1, num2 = take_input()
        division(num1, num2)
    elif choice == 0:
        print('Good bye!')
    else:
        print('wrong choice')
```

```python
def calculator():
    choice = -1
    while choice !=0:
        print('For addition, press 1\n')
        print('For subtraction, press 2\n')
        print('For multiplication, press 3\n')
        print('For division, press 4\n')
        print('For Exit, press 0\n')
        choice = int(input('Enter the choice : '))
        calculate(choice)
```

```python
calculator()
```
